**Removing column separator from columns**

The columns in the dataset make use of column separator for decimals, converting this to point.

```python
#converting columns from object to numeric

def convert_to_numeric(df):
    # Replace commas with periods and convert to numeric
    df.iloc[:, 4:] = df.iloc[:, 4:].replace(',', '.', regex=True).apply(pd.to_numeric)
    return df

list_of_dataframes = [fossil_c0_df1, fossil_ghg_df1]

# Apply the conversion function to each dataframe in the list
list_of_dataframes = [convert_to_numeric(df) for df in list_of_dataframes]
```

**Merging the datasets**

Merging all datasets for easier analysis, first by common categories

Creating common columns for fossil_c0_df2, fossil_c0_df3, fossil_c0_df4 and fossil_c0_df1 to prevent null values when dataframes are merged. This would be done for the GHG datasets as well

```python
# Define mappings for each DataFrame -CO2
mapping_df2 = {'Substance': 'CO2', 'Sector': 'co2_per_gdp'}
mapping_df3 = {'Substance': 'CO2', 'Sector': 'co2_per_capita'}
mapping_df4 = {'Substance': 'CO2', 'Sector': 'total_co2'}

# Apply mappings to create new columns
for df, mapping in zip([fossil_c0_df2, fossil_c0_df3, fossil_c0_df4], [mapping_df2, mapping_df3, mapping_df4]):
    df['Substance'] = mapping['Substance']
    df['Sector'] = mapping['Sector']
```

```python
# Define mappings for each DataFrame - GHG
mapping_df2 = {'Substance': 'GHG', 'Sector': 'ghg_per_gdp'}
mapping_df3 = {'Substance': 'GHG', 'Sector': 'ghg_per_capita'}
mapping_df4 = {'Substance': 'GHG', 'Sector': 'ghg_co2'}

# Apply mappings to create new columns
for df, mapping in zip([fossil_ghg_df2, fossil_ghg_df3, fossil_ghg_df4], [mapping_df2, mapping_df3, mapping_df4]):
    df['Substance'] = mapping['Substance']
    df['Sector'] = mapping['Sector']
```

```python
# Combining CO2 emissions DataFrames

co2_dataframes = [fossil_c0_df1, fossil_c0_df2, fossil_c0_df3, fossil_c0_df4]
combined_co2_df = pd.concat(co2_dataframes, ignore_index=True)

# Combining GHG emissions DataFrames
ghg_dataframes = [fossil_ghg_df1, fossil_ghg_df2, fossil_ghg_df3, fossil_ghg_df4]
combined_ghg_df = pd.concat(ghg_dataframes, ignore_index=True)
```
